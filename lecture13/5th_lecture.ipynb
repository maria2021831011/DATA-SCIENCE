# Method Overriding in OOP

```python
class phone:
    def __init__(self, brand, model, price):
        self.brand = brand
        self.model = model
        self._price = price
    def get_price(self):
        return self._price    

    def make_call(self, number):
        return f"Calling {number} from {self.brand} {self.model}."

    def send_message(self, number, message):
        return f"Sending message to {number}: {message}"
         
```

# Types of Inheritance in OOP

- Single Inheritance
- Multilevel Inheritance
- Hierachical Inheritance
- Multiple Inheritance

```python
#multilevel inheritance
```

```python
# Multilevel Inheritance


#Gradfather
class Product:
  def review(self):
    print("Product customer review")


#Father
class Phone(Product):
  def __init__(self, price, brand, camera):
    print("Inside Phone constructor")
    self.__price = price
    self.brand = brand
    self.camera = camera


  def buy(self):
    print("Buying a phone")


#child
class SmartPhone(Phone):
  pass



S  = SmartPhone(200000, "Samsung", 3)
S.buy()
S.review()
```

```python
# Hierarchical

#parent
class Phone:
  def __init__(self, price, brand, camera):
    print("Inside Phone constructor")
    self.__price = price
    self.brand = brand
    self.camera = camera


  def buy(self):
    print("Buying a phone")


#child
class SmartPhone(Phone):
  pass

#child
class FeaturePhone(Phone):
  pass



S = SmartPhone(100000, "apple", "3")
S.buy()
```

```python
F = FeaturePhone(100000, "apple", "3")
F.buy()
```

```python
# Mutiple

# parent
class Phone:
  def __init__(self, price, brand, camera):
    print("Inside Phone constructor")
    self.__price = price
    self.brand = brand
    self.camera = camera


  def buy(self):
    print("Buying a phone")

#parent
class Product:

  def __init__(self):
    pass

  def review(self):
    print("Product customer review")

#child
class SmartPhone(Phone, Product):
  pass




S = SmartPhone(100000, "apple", "3")
S.buy()
S.review()
```

Polymorphism in OOP
Polymorphism refers to having multiple forms. Polymorphism is a programming term that refers to the use of the same function name, but with different signature

```python
class Shape:

  def area(self,redius):
    print("this is 1st")
    return 3.14*redius**2

  def area(self, l,b):
    print("this is 2nd")
    return l*b


s = Shape()
s.area(2)
```

Abstraction in OOP
Abstraction is a fundamental concept in Python programming that allows us to simplify complex concepts and focus on the essential details. It involves hiding unnecessary details and exposing only the relevant information to the users

```python
# Abstract Base Class (ABC)
from abc import ABC, abstractmethod


#senior
class BankServer(ABC):
  def database(self):
    print("Connected to database")


  @abstractmethod
  def security(self):
    print("Security passed")



#Junior
class MobileApp(BankServer):
  def mobile_login(self):
        print("Mobile app login successful")


  def security():
    print("Security passed")
```

```python
mob=MobileApp()
```

```python
mob.mobile_login()
```
