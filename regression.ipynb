```python
import numpy as np
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
```

```python
# 1. Prepare Data (Actual vs Predicted)
y_actual = np.array([100, 90, 80, 70])
y_pred = np.array([98, 92, 78, 75])
```

```python
# 2. Calculate Absolute & Squared Metrics
mae = mean_absolute_error(y_actual, y_pred)
mse = mean_squared_error(y_actual, y_pred)
rmse = np.sqrt(mse)
```

```python
print(f"MAE: {mae}")
print(f"MSE: {mse}")
print(f"RMSE: {rmse}")
```

```python
# 3. Calculate R2 Score
r2 = r2_score(y_actual, y_pred)
```

```python
print(f"R2 Score: {r2}")
```

```python
# 4. Calculate Adjusted R2 (Manual Formula)
n = len(y_actual) # 4 observations
p = 1           # 1 feature (independent variable)
adj_r2 = 1 - (1 - r2) * (n - 1) / (n - p - 1)
```

```python
print(f"Adjusted R2: {adj_r2}")
```
