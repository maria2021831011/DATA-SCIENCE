Loop Control Statements
Break
Continue
Pass



```python
# break

for i in range(1,11):
    if i == 5:
        break
    print(i)
```

```python
name = ["Alex", "Mamum", "Bappy", "Ridoy", "Dabasish"]

for i in name:
    if i == "Bappy":
        print(i)
        break
```

```python
# continue

for i in range(1,10):
    if i == 5:
        continue
    print(i)
```

```python
# pass

for i in range(1,11):
    pass
```
