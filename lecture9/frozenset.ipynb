### Frozenset
Frozen set is just an immutable version of a Python set object

fs1 = frozenset([1,2,3,4])
fs2 = frozenset([3,4,5,6])
print(fs1)
print(fs2)

```python
fs1 = frozenset([1,2,3,4])
fs2 = frozenset([3,4,5,6])
print(fs1)
print(fs2)
```

```python
fs1 | fs2
```

```python
fs1 & fs2
```

```python
# 2D
fs = frozenset([1,2,frozenset([3,4])])
fs
```
