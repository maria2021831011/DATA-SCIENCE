### Difference between Lists and Tuples

- Syntax
- Mutability
- Speed
- Memory
- Built in functionality
- Error
- Usability

```python
import sys
t=list(range(1000))
p=tuple(range(1000))
print("size of list",sys.getsizeof(t))
print("size of tupple",sys.getsizeof(p))
```

```python
a = [1,2,3]
b = a
a.append(4)
print(a)
print(b)
```

```python
print(id(a))
print(id(b))
```

```python
a = (1,2,3)
b = a

a = a + (4,)
print(a)
print(b)
```
