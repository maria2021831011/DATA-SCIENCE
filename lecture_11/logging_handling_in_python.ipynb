```python
import logging
import os
```

```python
LOG_DIR="logs"
LOG_FILE_NAME="application.log"
os.makedirs(LOG_DIR,exist_ok=True)
log_path=os.path.join(LOG_DIR,LOG_FILE_NAME)
logging.basicConfig(
filename=log_path,
format="[%(asctime)s]%(name)s-%(levelname)s-%(message)s",
level=logging.INFO


)
```

```python
def profile(name):
    if type(name)==str:
        print(f"welcome {name}")
    else:
        logging.info("your input is not valid")    
```

```python
profile(1234)
```

```python
def devide_numbers(a,b):
    if b == 0:
        logging.info("Can't devide numbers with 0")
    else:
        return a / b
```

```python
devide_numbers(3,0)
```
