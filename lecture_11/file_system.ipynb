# File Handling in Python

##### Types of data used for I/O:
- Text - '12345' as a sequence of unicode chars
- Binary - 12345 as a sequence of bytes of its binary equivalent

##### Hence there are 2 file types to deal with
- Text files - All program files are text files
- Binary Files - Images,music,video,exe files



### How File I/O is done in most programming languages

- Open a file
- Read/Write data
- Close the file

```python
f = open("sample.txt", "w")
f.write("hello")
f.close()
```

```python
f.write("h")
```

```python
#multiple content akbare add
l=("hi\n iam maria\n")
f = open("sample.txt", "w")
f.writelines(l)
f.close()
```

```python
f = open("sample.txt", "r")
s=f.read()
print(s)
f.close()
```

```python
#to read line by line
f=open("sample.txt","r")
print(f.readline(),end='')
f.close()
```

```python
f = open("sample.txt", "r")
while True:
    data=f.readline()
    if data=='':
        break
    else:
        print(data,end="")
f.close()
```

# File Handling using Context Manager (With)
### Using Context Manager (With)

- It's a good idea to close a file after usage as it will free up the resources
- If we dont close it, garbage collector would close it
- with keyword closes the file as soon as the usage is over

```python
with open("sample.txt", "r") as f:
    s=f.read()
    print(s)
f.close()
```

```python
with open("sample.txt", "r") as f:
    s=f.readline()
    print(s)
f.close()
```

```python
with open("sample.txt", "a") as f:
    f.write(" i am so so")
f.close()
```

```python
with open("sample.txt", "r") as f:
    print(f.read(10))
    
f.close()
```

```python
with open("sample.txt", "r") as f:
    s=f.read()
    print(s)
f.close()
```

read a file

```python
with open("sample.txt", "r") as f:
    f.seek(8)
    print(f.read())
f.close()
```

```python
with open("sample.txt", "r") as f:
    f.seek(8)
    print(f.tell())
f.close()
```

```python
with open ("ccat.webp","r") as f:
    print(f.read())
```

```python
with open("ccat.webp",'rb') as f:
    print(f.read())
```

```python
#working with binary file
with open ("ccat.webp","rb")as f:
    original=f.read()
    with open("ccat_copy.webp","wb") as wf:
        wf.write(original)
```

```python
#working with other data type
with open("ccat.webp","w")as f:
    f.write(5)
```

```python
#working with other data type
with open("ccat.webp","w")as f:
    f.write("5")
```

```python
#working with other data type
with open("ccat.webp","r")as f:
    data=f.read()
    print(type(data))
```

Serialization and Deserialization
Serialization - process of converting python data types to JSON format
Deserialization - process of converting JSON to python data types
What is JSON?
The full form of JSON is JavaScript Object Notation.

```python
import json
l=[1,2,3,4]
with open('sample.txt','w') as f:
    json.dump(l,f)
```

```python
import json
l=[1,2,3,4]
with open('sample.txt','r') as f:
    data=json.load(f)
    print(data)
    print(type(data))
```

```python
import json
d={'name':"maria",
   'age ':20,
   'gender ':'male'
   }
with open('sample.txt','w') as f:
    json.dump(d,f,indent=4)

```

```python
with open("sample.txt",'r')as f:
    data=json.load(f)
    print(data)
    print(type(data))
```

### Pickling
`Pickling` is the process whereby a Python object hierarchy is converted into a byte stream, and `unpickling` is the inverse operation, whereby a byte stream (from a binary file or bytes-like object) is converted back into an object hierarchy.

```python
def dis():
    return "i am maria and i am a student"
d=dis()
```

```python
#pickle dump
import pickle
with open("fuunction.pkl","wb")as f:
    pickle.dump(d,f)
```

```python
import pickle
with open("fuunction.pkl",'rb')as f:
      p=pickle.load(f)
      print(p)
```
